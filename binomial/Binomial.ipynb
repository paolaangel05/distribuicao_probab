```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
%matplotlib inline

CINZA = "#9E9E9E"
CINZA_TEXTO = "#515151"
AZUL = "#003F5c"
```

##Lançamento de uma moeda 10 vezes.  


```python
# Lançamento de uma moeda 5 vezes
n = 10
p = 1/2

```

###Calcule a **Função Massa de Probabilidade**

```python
fig, ax = plt.subplots(figsize=(14
,6), dpi=72)

# x = np.arange(binom.ppf(0.00001, n, p),
#               binom.ppf(0.99, n, p))
x = np.arange(0,n+1)
ax.plot(x, binom.pmf(x, n, p), 'o', color=AZUL, ms=6)
ax.vlines(x, 0, binom.pmf(x, n, p), colors=AZUL, lw=5, alpha=0.6)

ax.spines[['top','right','left']].set_visible(False)

ax.set_yticks([])
ax.spines['bottom'].set_color(CINZA)

ax.set_xlabel("Número de CARAS", fontsize=13)
ax.set_xticks(range(n), labels=range(n))

for i in range(len(x)):
  y = binom.pmf(i, n, p)
  ax.annotate(f'{y:.4f}', #texto do rótulo
                xy=(i,y), #Ponto (x,y) do rótulo
                xytext=(4,5), #distãncia relativa do rótulo ao ponto onde ele foi definido
                textcoords='offset points',
                # color=CINZA_TEXTO,
                fontsize=13,
                fontweight='medium')


fig.text(0.5, 0.90,'Função Massa de Probabilidade da Binomial',fontweight='medium', fontsize=18, ha='center');
```

###Calcule a **Função de Distribuição Acumulada**

```python
fig, ax = plt.subplots(figsize=(14
,6), dpi=72)

# x = np.arange(binom.ppf(0.00001, n, p),
#               binom.ppf(0.99, n, p))
x = np.arange(0,n+1)
ax.step(x, binom.cdf(x, n, p), where='post', color=AZUL, ms=6)
#ax.vlines(x, 0, binom.cdf(x, n, p), colors=AZUL, lw=5, alpha=0.6)

ax.spines[['top','right','left']].set_visible(False)

ax.set_yticks([])
ax.spines['bottom'].set_color(CINZA)

ax.set_xlabel("Número de CARAS 'até' ", fontsize=13)
ax.set_xticks(range(n), labels=range(n))

for i in range(len(x)):
  y = binom.cdf(i, n, p)
  ax.annotate(f'{y:.4f}', #texto do rótulo
                xy=(i,y), #Ponto (x,y) do rótulo
                xytext=(3,3), #distãncia relativa do rótulo ao ponto onde ele foi definido
                textcoords='offset points',
                # color=CINZA_TEXTO,
                fontsize=13,
                fontweight='medium')


fig.text(0.5, 0.90,'Função de Distribuição Acumulada da Binomial',fontweight='medium', fontsize=18, ha='center');
```
