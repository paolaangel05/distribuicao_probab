## Binomial

A `distirbuição binomial` modela a **quantidade de sucessos** em um número fixo de tentativas independentes, onde cada tentativa tem **apenas dois resultados possíveis**: sucesso ou fracasso.

- $n$ é o número de tentativas independentes
- $p$ é a probabilidade de sucesso em *cada* tentativa
- A variável aleatória X é o número de sucessos em $n$ tentativas
- Função de probabilidade:
  <center>$P(X=k) = \displaystyle\binom{n}{k} p^k (1 - p)^{n - k}$, onde $k = 0,1,\dots,n$.


```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
%matplotlib inline

#Definindo cores para os gráficos
CINZA = "#9E9E9E"
CINZA_TEXTO = "#515151"
AZUL = "#003F5c"
```

## Lançamento de uma moeda 10 vezes

<p> Quando lançamos uma moeda honesta, a probabilidade de termos cara é de 0.5 e a probabilidade de termos coroa é 0.5. Ao lançarmos essa moeda 10 vezes e seja X o número de caras, calcule a expressão da distribuição de probabilidade da variável X e sua distribuição de probabilidade acumulada.</p>


<b>Definindo os parâmetros:</b>

```python
n = 10 # número de lançamentos
p = 1/2 # probabilidade de ser cara em um evento simples
```

### **Função Massa de Probabilidade**

```python
fig, ax = plt.subplots(figsize=(14,6), dpi=72)

# x = np.arange(binom.ppf(0.00001, n, p),binom.ppf(0.99, n, p))
x = np.arange(0,n+1)
ax.plot(x, binom.pmf(x, n, p), 'o', color=AZUL, ms=6)
ax.vlines(x, 0, binom.pmf(x, n, p), colors=AZUL, lw=5, alpha=0.6)

ax.spines[['top','right','left']].set_visible(False)

ax.set_yticks([])
ax.spines['bottom'].set_color(CINZA)

ax.set_xlabel("Número de CARAS", fontsize=13)
ax.set_xticks(range(n), labels=range(n))

for i in range(len(x)):
  y = binom.pmf(i, n, p)
  ax.annotate(f'{y:.4f}', #texto do rótulo
                xy=(i,y), #Ponto (x,y) do rótulo
                xytext=(4,5), #distãncia relativa do rótulo ao ponto onde ele foi definido
                textcoords='offset points',
                # color=CINZA_TEXTO,
                fontsize=13,
                fontweight='medium')


fig.text(0.5, 0.90,'Função Massa de Probabilidade da Binomial',fontweight='medium', fontsize=18, ha='center');
```

Como podemos observar, em uma moeda honesta, **a probabilidade de obter 50% de caras é a mais alta** dentre todos os resultados possíveis de número de caras, especialmente quando o número de lançamentos $n$ é par. Além disso, a maior probabilidade, ou seja, a `moda`, ocorre no valor de $k$ mais próximo de $n*p$.  
Logo, o valor mais provável de caras é $n*p = n/2 = 5$
